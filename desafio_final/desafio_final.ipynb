```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Dados-covid-19-municipios.csv', encoding= "ISO-8859-1", delimiter =';')
```

```python
df
```

```python
df = df[["Município","Mun_Total de casos","Mun_Total de óbitos"]]
df
```

```python
cidades = df.Município.value_counts(dropna=True)
cidades 
```

```python
casos = df[["Mun_Total de casos"]]
casos 

grupo = df['Mun_Total de casos'].value_counts(dropna=True)
grupo = grupo.to_frame()
casos_grupos = grupo.index
casos_grupos
```

```python
obtos = df[["Mun_Total de óbitos"]]

plt.plot(casos_grupos)
plt.ylabel('Maximo de casos - Covid') #legenda
plt.xlabel('Manimo de casos - Covid')
plt.show()

plt.plot(obtos)
plt.ylabel('Maximo de Obitos') #legenda
plt.xlabel('Minimo de Obitos') #legenda
plt.show()
```

```python
from fpdf import FPDF

pdf = FPDF ('P', 'mm', 'A4')

pdf.add_page()

pdf.set_font('Times','B',20)

texto = 'Grafico de Covid-019'
pdf.multi_cell(w=0,h=7,align='c', txt=texto)

pdf.image(name='grafico_casos.png', x=35, y=30, w=130)

pdf.image(name='grafico_obtos.png', x=35, y=170, w=130)
	
pdf.output('projeto.pdf')

print('Pdf criado')
```
